# Case study On Presentation of Data

## Operations on ‘cars_data’ dataset 

## 1.Read the data set to the python environment.

```python
# importing necessary libraries  
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
cars_data = pd.read_csv("cars_data.csv")
cars_data.head()
```

## 2.Check for the null values present in the dataset.

```python
cars_data.isnull()
```

```python
cars_data.isnull().sum()
```

## 3. Plot a bar graph of male vs female buyers participated in the sales.  

```python
plt.figure(figsize = (8,6))
plt.bar(cars_data['Buyer Gender'],height = 0.4,width = 0.5)
plt.title('Male Vs Female',fontsize = 16)
plt.xlabel('Gender',fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.show()
```

## 4. Find the top 5 cars based on their sales price.

```python
cars_data['Sale Price'].nlargest(n=5)
```

```python
cars_data['Sale Price'].head(5)
```

```python
cars_data1 = cars_data.nlargest(5,['Sale Price'])
cars_data1
```

```python
cars_data1[['Make','Model','Sale Price']]
```

## 5. Find the least 5 cars based on their Resell price. 

```python
cars_data['Resell Price'].nsmallest(n=5)
```

```python
cars_data['Resell Price'].tail(5)
```
